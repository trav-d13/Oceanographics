**Concept:**
Illustrate the concepts of a Taylor series expansion of $sin(x)$. Note that this will be a Mclauren series, based around point $x_0 = 0$. 

Establish the function of $sin(x)$

```python
x = np.linspace(-10, 10, 2000)
y = np.sin(x)
```

```python
plt.plot(x, y, 'k', linewidth=2)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid()
plt.title('sin(x)')
plt.show()
```

Plot the Mclaurin series expansions

```python
# Sin function
plt.plot(x, y, 'k', linewidth=2)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid()

# First order Taylor series expansion
P1 = [1, 0] # x + 0
yT1 = np.polyval(P1, x)
plt.plot(x, yT1, 'b--')

# Third order Taylor series expansion
P2 = [-1 / factorial(3), 0, 1, 0]
yT2 = np.polyval(P2, x)
plt.plot(x, yT2, 'r--')

# Fifth order Taylor series expansion
P3 = [1 / factorial(5), 0, -1 / factorial(3), 0, 1, 0]
yT3 = np.polyval(P3, x)
plt.plot(x, yT3, 'g--')

# Seventh order Taylor series expansion
P4 = [-1 / factorial(7), 0, 1 / factorial(5), 0, -1 / factorial(3), 0, 1, 0]
yT4 = np.polyval(P4, x)
plt.plot(x, yT4, 'y--')

# Nineth order Taylor series expansion
P5 = [1 / factorial(9), 0, -1 / factorial(7), 0, 1 / factorial(5), 0, -1 / factorial(3), 0, 1, 0]
yT5 = np.polyval(P5, x)
plt.plot(x, yT5, 'm--')

plt.legend(['sin(x)', '1st order', '3rd order', '5th order', '7th order', '9th order'])
plt.title('Taylor series expansion of sin(x) around x=0')
plt.show()
```
