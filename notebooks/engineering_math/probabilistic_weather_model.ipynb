# Probabilistic Weather Model
Based on the example provided in [*An Introduction to Modelling with Matrices and Vectors* by Steve Brunton](https://youtu.be/K-8F_zDMDUI?si=U0VJK-vRnSzrl6yy). 
All credit for the example goes to Steve Brunton, the original creator. Please refer to the original for accurate and original content. 



The below figure represents the transition between weather states. 
The values between weather states represent the probability of transitioning from the original state into the next state.



The system state $k$ can be represented as follows:

$$x_k = \begin{pmatrix} Pr(\text{Rain}) \\ Pr(\text{Nice}) \\ Pr(\text{Cloud})\end{pmatrix}$$

To define the next state, a transition matrix $A$ can be used. $A$ is defined as follows:

- The columns represent the starting states (for example column 1 represents starting state of Rain today)
- The rows represent the probability of moving from the starting state to a next state (for example position [0, 0] represent moving from a Rainy state today to a rain state tomorrow which is 0.5)

$$A = \begin{pmatrix} 
    0.5 & 0.5 & 0.25 \\ 
    0.25 & 0 & 0.25 \\ 
    0.25 & 0.5 & 0.5
\end{pmatrix}$$

The general solution to calculate a sequential state is:

$$\underline{x}_{k+1} = \underline{\underline{A}} \cdot \underline{x}_k$$

```python
A = np.array([[0.5, 0.5, 0.25], [0.25, 0.0, 0.25], [0.25, 0.5, 0.5]])
```

```python
print(A)
```

```python
x_today = np.array([[1], [0], [0]])
```

```python
print(x_today)
```

```python
# Calculate the subsequent state
print(A@x_today)
```

Showcase the iterative change in state probability over a long period of time.
This relates to steady states of systems. 

```python
range_limit = 50
x_today =np.array([[1], [0], [0]])  # Reset the original variable before executing
weather_store = np.zeros((3, range_limit))  # Store for changing weather states

for i in range(range_limit):
    weather_store[:, i] = x_today[:, 0]  # Save the current day state probability
    x_next = A@x_today  # Calculate next state weather probability
    x_today = x_next  # Transfer states from tomorrow to make it today
```

```python
plt.plot(weather_store.transpose())
plt.grid(True)
```

Note the existence of the link between steady states achieved in this example and Eigenvectors. 
Eigenvectors allow for calculation of steady state without iteration.
